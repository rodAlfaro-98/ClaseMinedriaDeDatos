```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Diccionario de Datos**

- ID: Id unico de cada canción en spotify
- Name: nombre de la canción
- Duration: Duración de la canción en minutos
- Energy: Representación percibida de la intensidad y actividad de la canción
- Key: Clave de la canción: [0=C, 1=C#/Db, 2=D, 3=D#/Eb, 4=E, 5=F, 6=F#/Gb, 7=G, 8=G#/Ab, 9=A, 10=A#/Bb, 11=B]
- Loudness: decibeles medidos a lo largo de toda la canción.
- Mode: Modalidad (Mayor, Menor)
- Speechiness: cantidad de palabras habladas en la canción.
- Acousticness: Medida entre 0.0 y 1.0 de si la canción es acústica
- INstrumentalness: Si una canción no presenta palabras.
- liveness: Emotividad de la canción
- valence: 
- tempo: Tiempo del compass utilizado en la canción
- danceability: Qué tan posible es bailar esa canción

# Paso 3: Detección de valores atipicos

**1) Distribución de variables numéricas**

```python
canciones.hist(figsize=(15,15),xrot=45)
plt.show()
```

Variables con datos atípicos:
1. Instrumentalness tiene casi el 100% de sus datos con un valor entre 0.0 y 0.05, mientras que existe un registro con 0.1 y otro con poco más de 0.4.
2. Liveliness tiene la más de laa mitad de datos entre 0.0 y 0.1, mientras que tiene un solo dato que se acerca mucho a 0.8

Fuera de eso se pueden observar las siguientes cosas:
- Duration tiene una distribución cuasinormal con ligero sesgo a la izquierda.
- Energy y Loudness tienen una distribución de campana de Gauss con sesgo a la derecha.
- Key, Valence, Tempo y Danceability realmente no siguen ningun tipo de distribución, parienciendo algunos a la distribución de generación de unos cuantos números aleatorios.
- Acousticness, Liveliness y Speechiness tienen un gran sesgo hacia la izquierda.

**3) Diagramas para detectar posibles valores atípicos**

```python
VariablesValoresAtipicos = ["duration","liveness","instrumentalness"]
for col in VariablesValoresAtipicos:
    sns.boxplot(data = canciones[col], orient = 'h').set_title(col)
    plt.show()
```

Checamos el resto de las variables

```python
VariablesValoresAtipicos = canciones.columns[2:len(canciones.columns)]
for col in VariablesValoresAtipicos:
    sns.boxplot(data = canciones[col], orient = 'h').set_title(col)
    plt.show()
```

Como podemos ver, las variables con valores atípicos son:
- Energy
- Duration
- Loudness
- Speechiness
- Instrumentalness
- Liveness

**4) Distribución de variables categóricas**

```python
canciones.describe(include='object')
```

**Plot**

Para este caso mostraremos la variable key, aunque esta variable es de tipo numérico realmente es una variable categórica que nos indica la llave en la que está escrita cada canción

```python
keys = canciones['key']
keys = keys.map({0:"C", 1:"C#/Db", 2:"D", 3:"D#/Eb", 4:"E", 5:"F", 6:"F#/Gb", 7:"G", 8:"G#/Ab", 9:"A", 10:"A#/Bb", 11:"B"})
sns.countplot(y=keys);
plt.show()
```

**5) Agrupación por variables categóricas**

```python
canciones['string_keys'] = canciones['key'].map({0:"C", 1:"C#/Db", 2:"D", 3:"D#/Eb", 4:"E", 5:"F", 6:"F#/Gb", 7:"G", 8:"G#/Ab", 9:"A", 10:"A#/Bb", 11:"B"})
display(canciones.groupby('string_keys').agg(['mean']))
```

# Paso 4: Identificación de relaciones entre pares de variables

```python
canciones.corr()
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(canciones.corr(), cmap='RdBu_r', annot=True)
plt.show()
```

```python
plt.figure(figsize=(14,7))
MatrizInf = np.triu(canciones.corr())
sns.heatmap(canciones.corr(), cmap='RdBu_r', annot=True, mask=MatrizInf)
plt.show()
```
